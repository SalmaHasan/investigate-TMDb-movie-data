### Data Cleaning (Replace this with more specific notes!)

```python
# copy dataframes for cleaning purposes
df_clean = df.copy()
```

```python

# Deleting columns will not use

df_clean.drop(['homepage', 
'keywords',
'status',               
'tagline'], inplace=True, axis=1)
```

```python
# Convert 'release_date' type from str to datetime 
df_clean["release_date"] = pd.to_datetime(df_clean["release_date"])
```

```python
# Replace all values of 0 with NAN
df_clean = df_clean.replace(0, np.nan)
```

```python
# Droping the null
df_clean = df_clean.dropna()
```

```python
df_clean.fillna(df_clean.mean(), inplace =True)
```

## Which year has the highest release of movies?

```python
# make group for each year and count the number of movies in each year 
data=df_clean.groupby('release_date').count()['id']

#make group of the data according to their release year and count the total number of movies in each year and pot.
df_clean.groupby('release_date').count()['id'].plot()

#set the figure size and labels
sns.set(rc={'figure.figsize':(10,10)})
plt.title("Year Vs Number Of Movies")
plt.xlabel('Release year')
plt.ylabel('Number Of Movies')
#set the style sheet
sns.set_style("whitegrid")
```

## What is the relation between number of movies and years?
We will work with two variables, The independent varible is the time in release_year column and the dependent varible is the number of movies.

First we need to calculate the number of movies in each year by value_counts() release_years column, and sort the years by sort_index().

Now we want visualize the result to make conclusion.

```python
# count the number of movies in each year then sort the result
count_movies_year = df["release_date"].value_counts().sort_index()

# plot relationship between release_year and number of movies
count_movies_year.plot(kind='barh',title='Number Of Movies vs Year', figsize=(5,10))

plt.xlabel('Number Of Movies')

plt.ylabel('Year')
```

## Which genre has the highest number of movies ?
First we need to define a function to separate genres then calculate the number of movies in each genre.


```python
 
def separate_count(column):
    split_data = pd.Series(df[column].str.cat(sep = '|').split('|'))
    
    count_data = split_data.value_counts(ascending=False)
    return count_data

```

```python
#Now we want visualize the result to make conclusion.

# Plot pie relationship between genre and number of movies
separate_count("genres").plot(kind="pie",figsize=(9,9))
# the title of the plot
plt.title('Percentage Of Genres')
plt.ylabel('');
```

## Which Movie Has The Highest Or Lowest Profit?


```python
#calculate Profit for each of the movie
#add a new column in the dataframe name 'Profit'
df_clean['Profit'] = df_clean['revenue'] - df_clean['budget']
```

```python
#use the function 'idmin' to find the index of lowest and highest profit movie.

#print the row related to this index.
def find_minmax(x):
    #use the function 'idmin' to find the index of lowest profit movie.
    min_index = df_clean[x].idxmin()
    #use the function 'idmax' to find the index of Highest profit movie.
    high_index = df_clean[x].idxmax()
    high = pd.DataFrame(df_clean.loc[high_index,:])
    low = pd.DataFrame(df_clean.loc[min_index,:])
    
    #print the movie with high and low profit
    print("The Highest Movie "+ x + " : ",df_clean['original_title'][high_index])
    print("The Lowest Movie "+ x + "  : ",df_clean['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

#call the find_minmax function.
find_minmax('Profit')
```

can conculde that Movie Which Has Highest Profit :  Avatar
Movie Which Has Lowest Profit  :  The Lone Ranger
